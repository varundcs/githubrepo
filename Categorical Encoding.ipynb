## Approach 2 : Label encoding

Another approach to encoding categorical values is to use a technique called label encoding. Label encoding is simply converting each value in a column to a number. 

One trick you can use in pandas is to convert a column to a category, then use those category values for your label encoding:

Label encoding has the advantage that it is straightforward but it has the disadvantage that the numeric values can be “misinterpreted” by the algorithms. For example, the value of 0 is obviously less than the value of 4 but does that really correspond to the data set in real life? 

```python
df_obj['body_style'] = df_obj['body_style'].astype('category')
```

```python
df_obj.dtypes
```

```python
df_obj["body_style_cat"] = df_obj["body_style"].cat.codes
df_obj
```

# Approach #3 - One Hot Encoding

A common alternative approach is called one hot encoding (but also goes by several different names shown below). Despite the different names, 

the basic strategy is to convert each category value into a new column and assigns a 1 or 0 (True/False) value to the column. This has the benefit of not weighting a value improperly but does have the downside of adding more columns to the data set.

Pandas supports this feature using get_dummies. This function is named this way because it creates dummy/indicator variables (aka 1 or 0).

```python
df_obj.drive_wheels.unique()
```

```python
df_obj['drive_wheels'].unique()
```

```python
pd.get_dummies(df_obj,columns= ['drive_wheels','body_style'], prefix= ['drive','body'])
```

```python
Please note that the get_dummies convert all the object type columns (categorial columns) .So be careful and use select_dtypes
```

# Scikit learn method to encode

 if we want to do a label encoding on the make of the car, we would import a LabelEncoder object and fit_transform on the data



```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
df_obj["make_code"] = lb_make.fit_transform(df_obj["make"])
df_obj[["make", "make_code"]].head(11)
```

```python
## Scikit-learn also supports binary encoding by using the LabelBinarizer. 
```

```python
from sklearn.preprocessing import LabelBinarizer
lb_fuel_system = LabelBinarizer()
lb_results = lb_fuel_system.fit_transform(df_obj['fuel_system'])

pd.DataFrame(lb_results,columns=lb_fuel_system.classes_)


```
